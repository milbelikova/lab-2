#### Шаг 2: Диаграммы рассеяния

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

##### b. Разделите данные.
Создайте два новых кадра данных: menDf и womenDf, каждый из которых содержит соответствующие записи.

```python
menDf = brainFrame[brainFrame['Gender'] == 'Male']
womenDf = brainFrame[brainFrame['Gender'] == 'Female']
menDf, womenDf
```

##### c. Постройте графики.

```python
# Ячейка для кода № 6
menMeanSmarts = menDf[['PIQ', 'FSIQ', 'VIQ']].mean(axis=1)
plt.scatter(menMeanSmarts, menDf['MRI_Count'])
plt.xlabel('Средний интеллект (PIQ, FSIQ, VIQ)')
plt.ylabel('MRI_Count')
plt.title('Мужчины: интеллект и размер мозга')
plt.show()
```

Аналогичным образом создайте график диаграммы рассеяния для кадра данных, содержащему только женские записи.

```python
# Ячейка для кода № 7
womenMeanSmarts = womenDf[['PIQ', 'FSIQ', 'VIQ']].mean(axis=1)
plt.scatter(womenMeanSmarts, womenDf['MRI_Count'])
plt.xlabel('Средний интеллект (PIQ, FSIQ, VIQ)')
plt.ylabel('MRI_Count')
plt.title('Женщины: интеллект и размер мозга')
plt.show()
```

## Часть 3: Вычисление корреляции с Python

#### Шаг 1: Рассчитайте корреляцию для brainFrame.

```python
brainFrame.corr(method='pearson', numeric_only=True)
```

Обратите внимание на диагональ слева направо в таблице корреляции, сгенерированной выше. Почему диагональ заполнена значениями 1? Это совпадение? Объясните.

Нет, это не совпадение. На диагонали каждая переменная сравнивается сама с собой, а корреляция величины с самой собой всегда равна 1, то есть это идеальная положительная линейная связь.

Продолжая смотреть на таблицу корреляции выше, обратите внимание, что значения зеркалируются; значения под диагональю имеют зеркальный аналог над ней. Это совпадение? Объясните.

Это тоже не совпадение. Матрица корреляции симметрична: корреляция X с Y всегда равна корреляции Y с X, поэтому значения выше и ниже диагонали одинаковые.

Используя тот же метод `corr()`, вычислите корреляцию переменных, содержащихся в кадре с женскими записями:

```python
womenDf.corr(method='pearson', numeric_only=True)
```

И то же самое можно сделать для переменных, содержащихся в кадре с мужскими записями:

```python
# Используйте corr() для расчёта критерия корреляции Пирсона для кадра данных с мужчинами
menDf.corr(method='pearson', numeric_only=True)
```

#### Шаг 2: Нарисуйте корреляционную тепловую карту.

```python
# Ячейка для кода № 12
import seaborn as sns

wcorr = womenDf.corr(numeric_only=True)
sns.heatmap(wcorr, annot=True, cmap='coolwarm')
# plt.savefig('attribute_correlations.png', bbox_inches='tight')
```

Аналогичным образом создайте и отобразите тепловую карту для кадра данных с мужчинами.

```python
# Ячейка для кода № 14
mcorr = menDf.corr(numeric_only=True)
sns.heatmap(mcorr, annot=True, cmap='coolwarm')
```

У многих пар переменных корреляция близка к нулю. Что это значит?

Это означает, что между такими переменными нет выраженной линейной зависимости или она очень слабая. При изменении одной переменной другая не показывает устойчивого линейного роста или уменьшения.

Зачем делать разделение по полу?

Разделение по полу нужно, чтобы сравнение было более корректным. Рост, вес и размер мозга у мужчин и женщин в среднем различаются, и если анализировать все вместе, пол может стать скрытым фактором, который искажает корреляции.

Какие переменные имеют более сильную корреляцию с размером мозга (MRI_Count)? Это ожидалось? Объясните.

Обычно более сильную положительную корреляцию с `MRI_Count` показывают рост (`Height`), вес (`Weight`) и в меньшей степени показатели интеллекта (`PIQ`, `FSIQ`, `VIQ`). Это в целом ожидаемо: физические размеры тела и размер мозга часто связаны сильнее, чем когнитивные тесты, потому что интеллект зависит не только от размера мозга.
